```python
import matplotlib.pyplot as plt

def plot_histograms(dataframe, bins=30, figsize=(50, 50)):
    ### plots histograms for every numerical feature of a dataset ### 
    num_columns = len(dataframe.columns)
    
    num_rows = (num_columns + 1) // 2
    num_cols = 2

    fig, axes = plt.subplots(num_rows, num_cols, figsize=figsize)
    
    axes = axes.flatten()
    
    for i, column in enumerate(dataframe.columns):
        ax = axes[i]
        
        dataframe[column].plot(kind='hist', bins=bins, ax=ax, edgecolor='black', color='darkviolet')
        
        ax.set_title(f'Histogram - {column}')
        ax.set_xlabel(column)
        ax.set_ylabel('Frequency')
        
        
```

```python
plot_histograms(cc)
plt.savefig('hist.png',bbox_inches = 'tight', pad_inches = 1)
```

We don't have any missing data, we don't have any duplicates, and our classes are perfectly balanced, as all things should be. It also does not look like we have many outliers. You can see there may be a few outliers in features V8, V21, and V22, but they seem to be very small in frequency and do not worry me.

Looks like the data was well-cleaned prior to being uploaded to Kaggle. The data was also scaled using PCA before upload.

I believe this was primarily done to ensure complete anonymity, given the sensitive nature of the data. Given the scaling, it seems pretty much impossible to infer what these features may have been prior to scaling (like time of transaction, location of transaction, things like that).
